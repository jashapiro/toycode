```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def dragon(i):
    if i == 0:
        return "R"
    else:
        last = dragon(i - 1)
        flip = "".join(["L" if t == "R" else "R" for t in last[::-1]])
        return last + "R" + flip
```

```python
dragon(4)
```

```python
def make_path(rl_string):
    path = np.matrix([[0],[0]])
    direction = np.matrix([[0],[1]])
    R_turn = np.matrix([[0, 1],[-1, 0]]) 
    L_turn = np.matrix([[0,-1],[ 1, 0]]) 
    for d in rl_string:
        path = np.append(path, path[:,-1] + direction, axis = 1)
        if d == "R":
            direction = R_turn * direction
        else:
            direction = L_turn * direction
    path = np.append(path, path[:,-1] + direction, axis = 1)    
    return path
```

```python
def dragonplot(i):
    dragon_path = make_path(dragon(i))
    plt.plot(*dragon_path.tolist())
    plt.axis('off')
    plt.axis('equal')
    plt.show()
```

```python
dragonplot(13)
```
